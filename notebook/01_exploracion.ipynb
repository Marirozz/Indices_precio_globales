# 01 - Exploración de Datos

Este notebook explora en el dataset:
- Valores nulos
- Tipo de datos
- Cantidad de filas y columnas
- Identificar valores unicos
- Clasificacion de las columnas

```python
from src.load_data import load_data

data_path = '../data/raw/Monitoreo_Precios_Promedios_Globales_RD.xlsx'
```

```python
df = load_data(data_path)
df.head()
```

```python
print('------------------ Null Values ------------------')
print(df.isnull().sum())

print('------------------ Data Types ------------------')
print(df.dtypes)
```

```python

print('------------------ Info ------------------')
print(df.info())

print('------------------ Describe ------------------')
print(df.describe())
```

```python
print('--------------------------- Columns --------------------------')
print(df.columns.tolist())

print('------------------ Unique Values in Column 1 ------------------')
print(df[df.columns[1]].unique())  

print('------------------ Duplicated Columns ------------------')
print(df.columns.duplicated().any())
```

```python
cols_info = ['Orden', 'Nombre', 'Unidad de medida']

cols_fecha = [col for col in df.columns if col not in cols_info]

total_ceros = (df[cols_fecha] == 0).sum().sum()

print(f"Total de ceros en columnas de fecha: {total_ceros}")
```
